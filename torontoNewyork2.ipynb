```python
def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        #print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            500, 
            100)
            
        # make the GET request
        #print(requests.get(url).json())
        group_data = requests.get(url).json().get("response").get('groups')
        
        if group_data:
            results = group_data[0].get('items')
        else:
            results=''
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v.get("venue").get("id"),
            v.get('venue').get('name'), 
            v.get('venue').get('location').get('lat'), 
            v.get('venue').get('location').get('lng'),  
            v.get('venue').get('categories')[0].get('name')) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue ID',
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

<h2>Get price, likes count and rating for each restaurat</h2>

```python
def getRestaurntDetails(names,IDs):
    
    details_list=[]
    for name,ID in zip(names,IDs):
        #print(name)
            
        # create the API request URL
        url="https://api.foursquare.com/v2/venues/{}?client_id={}&client_secret={}&v={}".format(
            ID,
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION         )
            
        # make the GET request
        
        group_data = requests.get(url).json()
        
        #print(group_data)
         #break
        if group_data.get("response"):
        
            #print(group_data.get("response"))
        
            
            if group_data.get("response").get("venue"):
                if group_data.get("response").get("venue").get("price"):
                    if group_data.get("response").get("venue").get("price").get("message"):
                        
                            details_list.append([(name, 
                                                  group_data.get("response").get('venue').get("price").get("message"),
                                                  group_data.get("response").get('venue').get("rating"),
                                                  group_data.get("response").get('venue').get("likes").get("count")
                                                 ) ])
                            

    
    details = pd.DataFrame([item for detail_list in details_list for item in detail_list])
    #details.head()
    #print(details)
    details.columns = ['Rest_name', 
                       'Price', 
                       'Rating' ,
                       'Likes Count'
                      ]
    
    return(details)
```

```python
# one hot encoding
toronto_rest_onehot=pd.DataFrame()
toronto_rest_onehot = pd.get_dummies(toronto_rest_details2[['Price']], prefix="", prefix_sep="")
toronto_rest_onehot['Rest_name'] = toronto_rest_details2['Rest_name'] 
fixed_columns = [toronto_rest_onehot.columns[-1]] + list(toronto_rest_onehot.columns[:-1])
toronto_rest_onehot = toronto_rest_onehot[fixed_columns]

toronto_rest_onehot = pd.merge(toronto_rest_onehot, toronto_rest_details2, on='Rest_name')
toronto_rest_onehot.head()
```

<h2>Normalization<h/2>

```python
from sklearn.preprocessing import StandardScaler

X = toronto_rest_onehot3.values[:,:]
X = np.nan_to_num(X)
cluster_dataset = StandardScaler().fit_transform(X)
cluster_dataset
```

</h2> Cluster</h2>

```python
kclusters=4

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(cluster_dataset)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:30]
```

```python
df5_toronto=pd.DataFrame()
df5_toronto["Cluster Labels"]=kmeans.labels_
df5_toronto=pd.concat([df5_toronto,toronto_rest_details2], axis=1)
df5_toronto
```

```python
df5_toronto.dropna(inplace=True)
df5_toronto.drop_duplicates(inplace=True)
df5_toronto=df5_toronto[np.isfinite(df5_toronto['Rating'])]
df5_toronto.head()
```

<h2> Visualize the clusters </h2>

```python
# create map
map_clusters = folium.Map(location=[43.651070, -79.347015], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, kclusters))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(df5_toronto['Venue Latitude'], df5_toronto['Venue Longitude'], df5_toronto['Rest_name'], df5_toronto['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[int(cluster)-2],
        fill=True,
        fill_color=rainbow[int(cluster)-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
